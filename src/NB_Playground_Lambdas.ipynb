Addendum: Lambdas
------------------------

A **lambda function** (sometimes called an [anonymous function](https://en.wikipedia.org/wiki/Anonymous_function) allows you to define a function (usually small) that does not need to be named. This can be helpful if the function only needs to be used a few times, or if the function is going to be used as an argument to another function (often called a "higher order function", for example: "map()").

## Lambdas in Python

The basic format of a lambda function in Python is:
$$\text{lambda } \text{ argument(s) } : \text{ expression}$$

Where $\text{lambda}$ is a **keyword** letting the interpreter know that this is a lambda function, $\text{argument(s)}$ consists of one or more **arguments**, or parameters, and $\text{expression}$ defines the operation to be performed (usually something relatively simple).

## Example (single argument)
Define a lambda function to double a number.

<u>Note:</u> Here we are assigning the lambda to a variable "double_num", but we don't have to!

```python
double_num = lambda x : x + x
print (double_num(5))
```

## Example (two arguments)
Define a lambda function to multiply two numbers.

```python
multiply = lambda x, y : x * y
print (multiply(3, 4))
```

## Example (using lambda as an argument)
Here we can use a lambda directly as an argument without assigning it to a variable.

```python
values = [1, 2, 3]
values_offset = map(lambda x : x + 3, values)
print(list(values_offset))
```

## Example (scope)
In some languages, the scope that a lambda has access to needs to be explicitly called out (e.g., C++). In Python, we have access to other variables in our function by default.

```python
offset = 4
values = [1, 2, 3]
values_offset = map(lambda x : x + offset, values)
print(list(values_offset))
```
